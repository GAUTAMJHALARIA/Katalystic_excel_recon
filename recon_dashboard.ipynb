```python
import gradio as gr
import pandas as pd
import math

ROWS_PER_PAGE = 15


def reconcile(summary_file, crm_file, tolerance):

    summary_df = pd.read_excel(summary_file.name)
    crm_df = pd.read_excel(crm_file.name)

    summary_df = summary_df.rename(columns={"Grand Total": "Grand_total_summary"})
    crm_df = crm_df.rename(columns={"Grand Total": "Grand_total_crm"})

    df = pd.merge(summary_df, crm_df, on="Folio Number", how="outer")

    # Difference
    df["Difference"] = (
        df["Grand_total_summary"].fillna(0)
        - df["Grand_total_crm"].fillna(0)
    )

    # Default status
    df["Status"] = "MATCH"

    # Missing cases
    df.loc[df["Grand_total_summary"].isna(), "Status"] = "Missing in Summary"
    df.loc[df["Grand_total_crm"].isna(), "Status"] = "Missing in CRM"

    # Apply dynamic tolerance
    df.loc[
        (df["Grand_total_summary"].notna()) &
        (df["Grand_total_crm"].notna()) &
        (df["Difference"].abs() >= tolerance),
        "Status"
    ] = "MISMATCH"

    # % Difference
    df["% Difference"] = (
        df["Difference"] /
        df["Grand_total_summary"].replace(0, pd.NA)
    ) * 100

    df["% Difference"] = df["% Difference"].round(2)

    # Sort by biggest difference first
    df = df.sort_values(by="Difference", key=abs, ascending=False)

    # KPIs
    total = len(df)
    matches = (df["Status"] == "MATCH").sum()
    mismatches = (df["Status"] == "MISMATCH").sum()
    missing_crm = (df["Status"] == "Missing in CRM").sum()
    missing_summary = (df["Status"] == "Missing in Summary").sum()
    match_percent = round((matches / total) * 100, 2) if total > 0 else 0

    # Manual verification table
    manual_df = df[[
        "Folio no.",
        "Grand_total_summary",
        "Grand_total_crm",
        "Difference",
        "% Difference",
        "Status"
    ]].copy()

    manual_df = manual_df[manual_df["Status"] == "MISMATCH"]

    output_file = "reconciliation_report.xlsx"
    df.to_excel(output_file, index=False)

    return (
        df,
        manual_df,
        total,
        matches,
        mismatches,
        missing_crm,
        missing_summary,
        f"{match_percent}%",
        output_file
    )


def filter_and_paginate(df, status, folio_search, page):

    if df is None or len(df) == 0:
        return pd.DataFrame(), "0 pages"

    if status != "All":
        df = df[df["Status"] == status]

    if folio_search:
        df = df[df["Folio Number"].astype(str).str.contains(folio_search)]

    total_pages = max(1, math.ceil(len(df) / ROWS_PER_PAGE))
    page = min(max(1, int(page)), total_pages)

    start = (page - 1) * ROWS_PER_PAGE
    end = start + ROWS_PER_PAGE

    return df.iloc[start:end].copy(), f"{total_pages} pages total"


with gr.Blocks(theme=gr.themes.Default()) as demo:

    gr.Markdown("## Reconciliation Dashboard")

    with gr.Row():
        summary_file = gr.File(label="Summary File")
        crm_file = gr.File(label="CRM File")
        tolerance_input = gr.Number(value=1, label="Tolerance (₹)")
        run_btn = gr.Button("Run Reconciliation", variant="primary")

    df_state = gr.State()

    gr.Markdown("### Summary")

    with gr.Row():
        total_card = gr.Number(label="Total Records")
        match_card = gr.Number(label="Matches")
        mismatch_card = gr.Number(label="Mismatches")
        miss_crm_card = gr.Number(label="Missing in CRM")
        miss_sum_card = gr.Number(label="Missing in Summary")
        match_percent_card = gr.Textbox(label="Match %")

    gr.Markdown("### Filters")

    with gr.Row():
        status_filter = gr.Dropdown(
            ["All", "MATCH", "MISMATCH", "Missing in CRM", "Missing in Summary"],
            value="All",
            label="Status"
        )
        folio_search = gr.Textbox(label="Search Folio Number")
        page_number = gr.Number(value=1, label="Page")

    gr.Markdown("### Full Dataset")

    table_output = gr.Dataframe(interactive=False)
    page_info = gr.Textbox(label="Pages")

    gr.Markdown("### Manual Verification (Mismatches Only)")

    manual_table_output = gr.Dataframe(interactive=False)

    download_output = gr.File(label="Download Full Excel Report")

    run_btn.click(
        reconcile,
        inputs=[summary_file, crm_file, tolerance_input],
        outputs=[
            df_state,
            manual_table_output,
            total_card,
            match_card,
            mismatch_card,
            miss_crm_card,
            miss_sum_card,
            match_percent_card,
            download_output
        ]
    )

    status_filter.change(
        filter_and_paginate,
        inputs=[df_state, status_filter, folio_search, page_number],
        outputs=[table_output, page_info]
    )

    folio_search.change(
        filter_and_paginate,
        inputs=[df_state, status_filter, folio_search, page_number],
        outputs=[table_output, page_info]
    )

    page_number.change(
        filter_and_paginate,
        inputs=[df_state, status_filter, folio_search, page_number],
        outputs=[table_output, page_info]
    )

demo.launch(share=True)
```

```python
import gradio as gr
import pandas as pd
import math

ROWS_PER_PAGE = 15


def reconcile(summary_file, crm_file, tolerance):

    try:
        summary_df = pd.read_excel(summary_file.name)
        crm_df = pd.read_excel(crm_file.name)
    except Exception as e:
        raise gr.Error(f"Error reading file: {e}")

    # Clean column names
    summary_df.columns = summary_df.columns.str.strip()
    crm_df.columns = crm_df.columns.str.strip()

    # Validate required columns
    required_cols = ["Folio Number", "Grand Total"]

    for col in required_cols:
        if col not in summary_df.columns:
            raise gr.Error(f"'{col}' missing in Summary file")
        if col not in crm_df.columns:
            raise gr.Error(f"'{col}' missing in CRM file")

    # Standardize internal names
    summary_df = summary_df.rename(columns={
        "Folio Number": "Folio",
        "Grand Total": "Grand_total_summary"
    })

    crm_df = crm_df.rename(columns={
        "Folio Number": "Folio",
        "Grand Total": "Grand_total_crm"
    })

    # Merge
    df = pd.merge(summary_df, crm_df, on="Folio", how="outer")

    # Difference
    df["Difference"] = (
        df["Grand_total_summary"].fillna(0)
        - df["Grand_total_crm"].fillna(0)
    )

    # Default status
    df["Status"] = "MATCH"

    # Missing
    df.loc[df["Grand_total_summary"].isna(), "Status"] = "Missing in Summary"
    df.loc[df["Grand_total_crm"].isna(), "Status"] = "Missing in CRM"

    # Tolerance rule
    df.loc[
        (df["Grand_total_summary"].notna()) &
        (df["Grand_total_crm"].notna()) &
        (df["Difference"].abs() >= tolerance),
        "Status"
    ] = "MISMATCH"

    # % Difference
    df["% Difference"] = (
        df["Difference"] /
        df["Grand_total_summary"].replace(0, pd.NA)
    ) * 100

    df["% Difference"] = df["% Difference"].round(2)

    # Sort biggest mismatches first
    df = df.sort_values(by="Difference", key=abs, ascending=False)

    # KPIs
    total = len(df)
    matches = (df["Status"] == "MATCH").sum()
    mismatches = (df["Status"] == "MISMATCH").sum()
    missing_crm = (df["Status"] == "Missing in CRM").sum()
    missing_summary = (df["Status"] == "Missing in Summary").sum()
    match_percent = round((matches / total) * 100, 2) if total > 0 else 0

    # Export
    output_file = "reconciliation_report.xlsx"
    df.to_excel(output_file, index=False)

    return (
        df,
        total,
        matches,
        mismatches,
        missing_crm,
        missing_summary,
        f"{match_percent}%",
        output_file
    )


def filter_and_paginate(df, status, folio_search, page):

    if df is None or len(df) == 0:
        return pd.DataFrame(), pd.DataFrame(), "0 pages"

    # Apply status filter
    if status != "All":
        df = df[df["Status"] == status]

    # Apply folio search
    if folio_search:
        df = df[df["Folio"].astype(str).str.contains(folio_search)]

    total_pages = max(1, math.ceil(len(df) / ROWS_PER_PAGE))
    page = min(max(1, int(page)), total_pages)

    start = (page - 1) * ROWS_PER_PAGE
    end = start + ROWS_PER_PAGE

    paginated_df = df.iloc[start:end].copy()

    # Manual verification table (same filtered dataset, fewer columns)
    manual_df = paginated_df[[
        "Folio",
        "Grand_total_summary",
        "Grand_total_crm",
        "Difference",
        "% Difference",
        "Status"
    ]].copy()

    return paginated_df, manual_df, f"{total_pages} pages total"


with gr.Blocks(theme=gr.themes.Default()) as demo:

    gr.Markdown("## Reconciliation Dashboard")

    with gr.Row():
        summary_file = gr.File(label="Summary File")
        crm_file = gr.File(label="CRM File")
        tolerance_input = gr.Number(value=1, label="Tolerance")
        run_btn = gr.Button("Run Reconciliation", variant="primary")

    df_state = gr.State()

    gr.Markdown("### Summary")

    with gr.Row():
        total_card = gr.Number(label="Total Records")
        match_card = gr.Number(label="Matches")
        mismatch_card = gr.Number(label="Mismatches")
        miss_crm_card = gr.Number(label="Missing in CRM")
        miss_sum_card = gr.Number(label="Missing in Summary")
        match_percent_card = gr.Textbox(label="Match %")

    gr.Markdown("### Filters")

    with gr.Row():
        status_filter = gr.Dropdown(
            ["All", "MATCH", "MISMATCH", "Missing in CRM", "Missing in Summary"],
            value="All",
            label="Status"
        )
        folio_search = gr.Textbox(label="Search Folio")
        page_number = gr.Number(value=1, label="Page")

    gr.Markdown("### Full Dataset")

    table_output = gr.Dataframe(interactive=False)
    page_info = gr.Textbox(label="Pages")

    gr.Markdown("### Manual Verification (Same Filtered Data)")

    manual_table_output = gr.Dataframe(interactive=False)

    download_output = gr.File(label="Download Full Excel Report")

    # Run reconciliation
    run_btn.click(
        reconcile,
        inputs=[summary_file, crm_file, tolerance_input],
        outputs=[
            df_state,
            total_card,
            match_card,
            mismatch_card,
            miss_crm_card,
            miss_sum_card,
            match_percent_card,
            download_output
        ]
    )

    # Apply filters to BOTH tables
    status_filter.change(
        filter_and_paginate,
        inputs=[df_state, status_filter, folio_search, page_number],
        outputs=[table_output, manual_table_output, page_info]
    )

    folio_search.change(
        filter_and_paginate,
        inputs=[df_state, status_filter, folio_search, page_number],
        outputs=[table_output, manual_table_output, page_info]
    )

    page_number.change(
        filter_and_paginate,
        inputs=[df_state, status_filter, folio_search, page_number],
        outputs=[table_output, manual_table_output, page_info]
    )

demo.launch(share=True)
```
