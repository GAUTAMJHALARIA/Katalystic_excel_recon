```python
import gradio as gr
import pandas as pd
import math

ROWS_PER_PAGE = 15


# =========================================================
# Payment State Logic
# =========================================================

def determine_summary_payment_state(balance, tolerance):
    if pd.isna(balance):
        return "Unknown"
    return "Closed" if abs(balance) <= tolerance else "Open"


def determine_erp_payment_state(status_series):
    """
    If ANY ERP status under a folio is 'Paid' → Closed
    Otherwise → Open
    """
    statuses = status_series.dropna().astype(str).str.lower()

    if len(statuses) == 0:
        return "Unknown"

    if "paid" in statuses.values:
        return "Closed"

    return "Open"


# =========================================================
# Reconciliation Logic
# =========================================================

def reconcile(summary_file, erp_file, tolerance):

    summary_df = pd.read_excel(summary_file.name)
    erp_df = pd.read_excel(erp_file.name)

    summary_df.columns = summary_df.columns.str.strip()
    erp_df.columns = erp_df.columns.str.strip()

    summary_df = summary_df.rename(columns={
        "Folio Number": "Folio",
        "Grand Total": "Grand_total_summary",
        "Balance": "Balance"
    })

    erp_df = erp_df.rename(columns={
        "Folio Number": "Folio",
        "Grand Total": "Grand_total_erp",
        "Status": "ERP_Status"
    })

    # =========================================================
    # Aggregate Summary per Folio
    # =========================================================

    summary_grouped = (
        summary_df.groupby("Folio", as_index=False)
        .agg({
            "Grand_total_summary": "sum",
            "Balance": "sum"
        })
    )

    # =========================================================
    # Aggregate ERP per Folio
    # =========================================================

    erp_grouped = (
        erp_df.groupby("Folio", as_index=False)
        .agg({
            "Grand_total_erp": "sum",
            "ERP_Status": lambda x: determine_erp_payment_state(x)
        })
    )

    # =========================================================
    # Merge (Now One Row Per Folio)
    # =========================================================

    df = pd.merge(summary_grouped, erp_grouped, on="Folio", how="outer")

    # =========================================================
    # Missing Detection
    # =========================================================

    df["Missing_in_Summary"] = df["Grand_total_summary"].isna()
    df["Missing_in_ERP"] = df["Grand_total_erp"].isna()
    df["Missing_Record"] = df["Missing_in_Summary"] | df["Missing_in_ERP"]

    # =========================================================
    # Amount Comparison
    # =========================================================

    df["Difference"] = (
        df["Grand_total_summary"].fillna(0)
        - df["Grand_total_erp"].fillna(0)
    )

    df["Amount_Mismatch"] = (
        df["Difference"].abs() >= tolerance
    ) & (~df["Missing_Record"])

    # =========================================================
    # Payment Comparison
    # =========================================================

    df["Summary_Payment_State"] = df["Balance"].apply(
        lambda x: determine_summary_payment_state(x, tolerance)
    )

    # ERP_Status already aggregated to Closed/Open
    df["ERP_Payment_State"] = df["ERP_Status"]

    df["Payment_Mismatch"] = (
        df["Summary_Payment_State"] != df["ERP_Payment_State"]
    ) & (~df["Missing_Record"])

    # =========================================================
    # Sort by Severity
    # =========================================================

    df = df.sort_values(
        by=["Missing_Record", "Amount_Mismatch",
            "Payment_Mismatch", "Difference"],
        ascending=[False, False, False, False]
    )

    # =========================================================
    # KPIs
    # =========================================================

    total = len(df)
    missing_count = df["Missing_Record"].sum()
    amount_mismatch = df["Amount_Mismatch"].sum()
    payment_mismatch = df["Payment_Mismatch"].sum()
    both_mismatch = (
        df["Amount_Mismatch"] & df["Payment_Mismatch"]
    ).sum()

    perfect_match = (
        (~df["Missing_Record"]) &
        (~df["Amount_Mismatch"]) &
        (~df["Payment_Mismatch"])
    ).sum()

    match_percent = round(
        (perfect_match / total) * 100, 2
    ) if total > 0 else 0

    output_file = "reconciliation_report.xlsx"
    df.to_excel(output_file, index=False)

    return (
        df,
        1,
        total,
        missing_count,
        amount_mismatch,
        payment_mismatch,
        both_mismatch,
        f"{match_percent}%",
        output_file
    )


# =========================================================
# Filtering Logic
# =========================================================

def filter_data(df, filter_type):

    if filter_type == "Missing Records":
        return df[df["Missing_Record"]]

    if filter_type == "Perfect Match":
        return df[
            (~df["Missing_Record"]) &
            (~df["Amount_Mismatch"]) &
            (~df["Payment_Mismatch"])
        ]

    if filter_type == "Amount Mismatch (Any)":
        return df[df["Amount_Mismatch"]]

    if filter_type == "Payment Mismatch (Any)":
        return df[df["Payment_Mismatch"]]

    if filter_type == "Both Issues":
        return df[
            df["Amount_Mismatch"] & df["Payment_Mismatch"]
        ]

    return df


def filter_and_paginate(df, filter_type, folio_search, current_page):

    if df is None or len(df) == 0:
        return pd.DataFrame(), pd.DataFrame(), "Showing 0–0 of 0", current_page

    filtered_df = filter_data(df, filter_type)

    if folio_search:
        filtered_df = filtered_df[
            filtered_df["Folio"].astype(str).str.contains(folio_search)
        ]

    total_rows = len(filtered_df)
    total_pages = max(1, math.ceil(total_rows / ROWS_PER_PAGE))
    current_page = max(1, min(current_page, total_pages))

    start = (current_page - 1) * ROWS_PER_PAGE
    end = min(start + ROWS_PER_PAGE, total_rows)

    page_df = filtered_df.iloc[start:end].copy()

    display_text = f"Showing {start + 1}–{end} of {total_rows}"

    return page_df, page_df.copy(), display_text, current_page


def next_page(df, filter_type, folio_search, current_page):
    return filter_and_paginate(df, filter_type, folio_search, current_page + 1)


def prev_page(df, filter_type, folio_search, current_page):
    return filter_and_paginate(df, filter_type, folio_search, current_page - 1)


def apply_filters(df, filter_type, folio_search):
    return filter_and_paginate(df, filter_type, folio_search, 1)


# =========================================================
# UI
# =========================================================

with gr.Blocks(theme=gr.themes.Default()) as demo:

    gr.Markdown("## Invoice Reconciliation Dashboard (Production Version)")

    with gr.Row():
        summary_file = gr.File(label="Bill Summary File")
        erp_file = gr.File(label="Sales Invoice (ERP) File")
        tolerance_input = gr.Number(value=1, label="Tolerance (₹)")
        run_btn = gr.Button("Run Reconciliation", variant="primary")

    df_state = gr.State()
    page_state = gr.State(1)

    gr.Markdown("### Summary")

    with gr.Row():
        total_card = gr.Number(label="Total Invoices")
        missing_card = gr.Number(label="Missing Records")
        amount_mismatch_card = gr.Number(label="Amount Mismatch")
        payment_mismatch_card = gr.Number(label="Payment Mismatch")
        both_mismatch_card = gr.Number(label="Both Issues")
        match_percent_card = gr.Textbox(label="Perfect Match %")

    gr.Markdown("### Filters")

    with gr.Row():
        filter_dropdown = gr.Dropdown(
            ["All",
             "Missing Records",
             "Perfect Match",
             "Amount Mismatch (Any)",
             "Payment Mismatch (Any)",
             "Both Issues"],
            value="All",
            label="Filter Type"
        )
        folio_search = gr.Textbox(label="Search Folio")

    gr.Markdown("### Results")

    table_output = gr.Dataframe(interactive=False)

    with gr.Row():
        prev_btn = gr.Button("◀ Previous")
        next_btn = gr.Button("Next ▶")

    count_display = gr.Textbox(label="Rows", interactive=False)
    download_output = gr.File(label="Download Full Excel Report")

    run_btn.click(
        reconcile,
        inputs=[summary_file, erp_file, tolerance_input],
        outputs=[
            df_state,
            page_state,
            total_card,
            missing_card,
            amount_mismatch_card,
            payment_mismatch_card,
            both_mismatch_card,
            match_percent_card,
            download_output
        ]
    )

    filter_dropdown.change(
        apply_filters,
        inputs=[df_state, filter_dropdown, folio_search],
        outputs=[table_output, table_output, count_display, page_state]
    )

    folio_search.change(
        apply_filters,
        inputs=[df_state, filter_dropdown, folio_search],
        outputs=[table_output, table_output, count_display, page_state]
    )

    next_btn.click(
        next_page,
        inputs=[df_state, filter_dropdown, folio_search, page_state],
        outputs=[table_output, table_output, count_display, page_state]
    )

    prev_btn.click(
        prev_page,
        inputs=[df_state, filter_dropdown, folio_search, page_state],
        outputs=[table_output, table_output, count_display, page_state]
    )

demo.launch(share=True)
```
